```python
# Aykr deerleri tespit etmek iin kutu grafii iz
sns.boxplot(data=sonar_data.iloc[:, :-1])  # Son stun (class) hari tm stunlar al
plt.title('Box Plot of Features')
plt.show()
```

```python
# Aykr deerleri filtrelemek iin IQR kullan
Q1 = sonar_data.iloc[:, :-1].quantile(0.25)
Q3 = sonar_data.iloc[:, :-1].quantile(0.75)
IQR = Q3 - Q1

# Aykr deerleri iermeyen bir veri seti olutur
sonar_data_filtered = sonar_data[~((sonar_data.iloc[:, :-1] < (Q1 - 1.5 * IQR)) | (sonar_data.iloc[:, :-1] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

### training with Logistic Regression

```python
# separating data and Labels
X = sonar_data_filtered.drop(columns=60, axis=1)
Y = sonar_data_filtered[60]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
#stratify=Y ifadesi, veri setinin blnmesi srasnda snflar arasndaki oranlarn korunmasn salamak iin kullanlr
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

### Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
print('Accuracy on test data : ', test_data_accuracy)
```
